# 2> Game


```python
# We have these kinds of messy data inside Game columns so clean them  - >  '', '  ', ., :, /, &, +, -, ', (2015), !, 

import numpy as np
df['Game'] = (
    df['Game']
    .astype('string')
    .str.strip()
    .str.lower()
    .str.replace(r"[^a-zA-Z0-9\s]", '', regex=True)    # keeps only letters,numbers
    .str.replace(r'\s+', ' ', regex=True)              # collapse multiple spaces
    .str.strip()                                       # strip
    .replace(r'^\s*$', np.nan, regex=True)
    .str.title()                                       # capitalize words
)
df['Game'].sample(10)
```

# 3> Platform ->

```python
df['Platform'].unique()
```

```python
df['Platform']=(df['Platform']
    .str.strip()
    .str.lower()
    .replace(r'^\s*$', np.nan, regex=True)
    .str.upper()
    .astype('category')
               )

df['Platform'].sample(10)
```

# 4> Genre

```python
df['Genre'].unique()
```

```python
# array(['Sports', 'Platform', 'Racing', 'Role-Playing', 'Puzzle', 'Misc',
#        'Shooter', 'Simulation', 'Action', 'Fighting', 'Adventure',
#        'Strategy'], dtype=object)

#  genre object o - c = Role-Playing -> Role Playing

df['Genre']=(df['Genre']
             .astype('string')
             .str.strip()
             .str.lower()
             .str.replace('-', ' ', regex=True)
             .replace(r'^\s*$', np.nan, regex=True)
             .str.title()
             .astype('category')
            )

# ['Sports', 'Platform', 'Racing', 'Role Playing', 'Puzzle', ..., 'Simulation', 'Action', 'Fighting', 'Adventure', 'Strategy']
# Length: 12
# Categories (12, string): [Action, Adventure, Fighting, Misc, ..., Shooter, Simulation, Sports, Strategy]


# valid_gerne=['Sports', 'Platform', 'Racing', 'Role Playing', 'Puzzle', 'Misc',
#        'Shooter', 'Simulation', 'Action', 'Fighting', 'Adventure',
#        'Strategy']

# df.loc[~df['Genre'].isin(valid_gerne), 'Genre']=np.nan

df['Genre'].sample(15)
```

```python
valid_gerne=['Sports', 'Platform', 'Racing', 'Role Playing', 'Puzzle', 'Misc',
       'Shooter', 'Simulation', 'Action', 'Fighting', 'Adventure',
       'Strategy']
```

```python
df.loc[~df['Genre'].isin(valid_gerne), 'Genre']=np.nan

df['Genre'].sample(15)
```

# 5> Publisher

```python
# Publisher object - [-.,] [N/A, Unknown]

df['Publisher'].unique()
```

```python
df['Publisher']=(df['Publisher']
                .astype('string')
                 .str.strip()
                 .str.lower()
                 .replace(r'[^a-zA-A0-9\s]', ' ', regex=True)
                 .replace(r'\s+', ' ', regex=True)
                 .replace(r'^\s*$', np.nan,regex=True)
                 .replace(['n/a','unknown'], np.nan)
                 .str.strip()
                 .str.title()
                 .astype('category')
                )

df['Publisher'].sample(10)
```

# 6> Year

```python
# Year int64 - we dont need change any values because no null values and data type is int64 so it cleaned

df['Year'].sample(20)
```

# 7> Month

```python
df['Month'].unique()
```

```python
# pd.set_option('display.max_rows',None)
df['Month'] # Eg:- index 268 see October , index 3767 see Novermber so change it to Oct,Nov
```

```python
# array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
#        'Oct', 'Nov', 'Dec', 'October', 'November'], dtype=object)

# o-c :-  october,november -> ,Oct, Nov
df['Month']=(df['Month']
    .astype('string')
    .str.strip()
    .str.lower()
    .replace({'october':'oct', 'november':'nov'})
    .str.capitalize()
    .astype('category')
            )

# ['Jan', 'Feb', 'Mar', 'Apr', 'May', ..., 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# Length: 12
# Categories (12, string): [Apr, Aug, Dec, Feb, ..., May, Nov, Oct, Sep]

# valid_month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
#        'Oct', 'Nov', 'Dec']

# df.loc[~df['Month'].isin(valid_month), 'Month']=np.nan

df['Month'].sample(10)
# Eg:- index 268 see October , index 3767 see Novermber so change it to Oct,Nov
```

```python
# pd.reset_option('display.max_rows')
```

```python
valid_month=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
       'Oct', 'Nov', 'Dec']
```

```python
df.loc[~df['Month'].isin(valid_month), 'Month']=np.nan
df['Month'].sample(10)
```

# 8> Country

```python
df['Country'].unique()
```

```python
# Here i am we have 3 countries like USA, United States, Australia so replace United States make as USA

df['Country']=(df['Country']
               .astype('string')
               .str.strip()
               .replace({'United States':'USA'}, regex=True)
               .str.strip()
               .replace(r'^\s*$',np.nan, regex=True)
               .astype('category')
)
df['Country'].sample(20)
```

```python
valid_country=['USA','Australia']
```

```python
df.loc[~df['Country'].isin(valid_country), 'Country']=np.nan
df['Country'].sample(10)
```

# 9> State

```python
df['State'].unique()
```

```python
df['State']=(df['State']
            .astype('string')
            .str.strip()
            .str.title()
            .replace(r'^\s*$',np.nan, regex=True)
            .astype('category')
            )
df['State'].sample(10)
```

# 10> City

```python
df['City'].unique()
```

```python
df['City']=(df['City']
            .astype('string')
            .str.strip()
            .str.lower()
            .str.replace('-',' ', regex=True)
            .str.replace(r'\s+', ' ', regex=True)
            .str.strip()
            .str.title()
            .replace({'Mcallen':'McAllen'})
            .replace(r'^\s*$', np.nan, regex=True)
            .astype('category')
            
           )
df['City'].sample(10)
```

# 11> Region

```python
df['Region'].unique()
```

```python
df['Region']=(df['Region']
              .astype('string')
              .str.strip()
              .str.lower()
              .replace({'nan':np.nan},regex=True)
              .replace(r'^\s*$', np.nan, regex=True)
              .str.title()
              .astype('category')

)
df['Region'].sample(10)
```

# 12> Sale

```python
df['Sale'].unique()
```

```python
df['Sale']=pd.to_numeric(df['Sale']
               .astype('string')
               .str.strip()
               .str.replace(r'[^\d.]','', regex=True),
                errors='coerce')
df['Sale'].sample(5)
```

```python
df.loc[ (df['Sale']<0) | (df['Sale']>100), 'Sale']=np.nan
df['Sale'].sample(10)
```

# 13> Profit

```python
df['Profit'].unique()
```

```python
df['Profit']=(pd.to_numeric(df['Profit']
    .replace(r'[\d.]', '', regex=True),
    errors='coerce').round(2)
)
df['Profit'].sample(5)
```

```python
df.loc[ (df['Profit']<0) | (df['Profit']>100), 'Profit']=np.nan
df['Profit'].sample(10)
```
